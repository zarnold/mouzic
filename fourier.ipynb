```python
from math import pi, e, cos, sin
```

```python
import numpy as np 
import pylab as plt
plt.rcParams['figure.figsize'] = 29.7,12
time = np.arange(0,0.01, 1/44000)
```

```python
freqFondamental = 440
periodFondamental=1/freqFondamental
pulsationFondamental = 2*pi*freqFondamental

def harmonique(freqFond, nth):
    return np.cos(2*pi*time*nth*freqFondamental)


def generateHarmonic(freqFondamental, N):
    h=np.zeros((N,len(time)))
    for i in range(N):
        h[i] = harmonique(freqFondamental,i+1)
    return h
```

```python
def sumHarmonics(f,N, coef):
    H=generateHarmonic(f,N)
    e=np.dot(coef,H)
    return e
```

```python
c=np.array([0,1,0.4,0.8,0.1,0.3,0.6,0.2,0.1,0.1,0.05])
```

```python
f=sumHarmonics(440,len(c), c)
```

```python
from numpy.fft import fft, ifft, fftfreq
plt.plot(time,f)
```
